```python
# Libs for Data Treatment and Vizualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 0. Dataset Global Models (Sorocaba-Ipero 2024 | Server Side)⭐

```python
# ------------------------
# Data collection
# ------------------------

file_path2023 = 'Datasets tratados/2023/Processed_Data_2023-A713_Sorocaba_Ipero.csv'
file_path2024 = 'Datasets tratados/2024/Processed_Data_2024-A713_Sorocaba_Ipero.csv'

df_2023 = pd.read_csv(file_path2023,
                      encoding='utf-8',
                      index_col=0,
                      parse_dates=True)
df_2024 = pd.read_csv(file_path2024,
                      encoding='utf-8',
                      index_col=0,
                      parse_dates=True)
```

```python
# Verificar o tipo do índice
print("Tipo do índice:", type(df_2023.index))
print("É DatetimeIndex?", isinstance(df_2023.index, pd.DatetimeIndex))

print("\n<------------------------>\n")

if not isinstance(df_2023.index, pd.DatetimeIndex):
    print("Convertendo índice para datetime...")
    df_2023.index = pd.to_datetime(df_2023.index)
    print("Novo tipo:", type(df_2023.index))

df_2023.head()
```

```python
# Verificar o tipo do índice
print("Tipo do índice:", type(df_2024.index))
print("É DatetimeIndex?", isinstance(df_2024.index, pd.DatetimeIndex))

print("\n<------------------------>\n")

if not isinstance(df_2024.index, pd.DatetimeIndex):
    print("Convertendo índice para datetime...")
    df_2024.index = pd.to_datetime(df_2024.index)
    print("Novo tipo:", type(df_2024.index))

df_2024.head()
```

```python
# ----------------------------
# Data Similarity Verification
# ---------------------------

print("2023 DataFrame Shape:", df_2023.shape)
print("2024 DataFrame Shape:", df_2024.shape)
print("<-------------------------------->\n")

columns_2023 = list(df_2023.columns)
columns_2024 = list(df_2024.columns)
print("2023 DataFrame Columns:", columns_2023)
print("2024 DataFrame Columns:", columns_2024)
print("<-------------------------------->\n")

if columns_2023 == columns_2024:
    print("Both DataFrames have the same columns.")
else:
    print("DataFrames have different columns.")
    print("2023 Columns:", columns_2023)
    print("2024 Columns:", columns_2024)
```

```python
# ----------------------------
# Data Concatenation
# ---------------------------

df_concat = pd.concat([df_2023, df_2024])

df_concat["2023-12-31 20:00:00":"2024-01-01 2:00:00"]
```

```python
# ----------------------------
# Data Concatenation (Visualization in Graph)
# ---------------------------
df_concat_period = df_concat["2023-12-31 20:00:00":"2024-01-01 05:00:00"]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 4), sharex=True)

virada_ano = pd.Timestamp("2024-01-01 00:00:00")

for i, (periodo, cor, label_suffix) in enumerate([
    (df_concat_period.index < virada_ano, 'red', ''),
    (df_concat_period.index >= virada_ano, '#FF6B6B', ' (2024)')
]):
    mask = periodo
    if any(mask):
        ax1.plot(df_concat_period.index[mask], 
                 df_concat_period.loc[mask, 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], 
                 label=f'Temperatura{label_suffix}', 
                 color=cor, 
                 linewidth=2.5,
                 alpha=0.9,
                 marker='o',
                 markersize=3)

ax1.set_ylabel('Temperatura (°C)', fontsize=11, fontweight='bold')
ax1.legend(loc='best')
ax1.grid(True, alpha=0.2)
ax1.set_facecolor('#f8f9fa')
ax1.axvline(x=virada_ano, color='gray', linestyle='--', alpha=0.7)

for i, (periodo, cor, label_suffix) in enumerate([
    (df_concat_period.index < virada_ano, 'green', ''),
    (df_concat_period.index >= virada_ano, '#4ECDC4', ' (2024)')
]):
    mask = periodo
    if any(mask):
        ax2.plot(df_concat_period.index[mask], 
                 df_concat_period.loc[mask, 'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'], 
                 label=f'Pressão{label_suffix}', 
                 color=cor, 
                 linewidth=2.5,
                 alpha=0.9,
                 marker='s',
                 markersize=3)

ax2.set_xlabel('Data e Hora', fontsize=12, fontweight='bold')
ax2.set_ylabel('Pressão (mB)', fontsize=11, fontweight='bold')
ax2.legend(loc='best')
ax2.grid(True, alpha=0.2)
ax2.set_facecolor('#f8f9fa')
ax2.tick_params(axis='x', rotation=45)
ax2.axvline(x=virada_ano, color='gray', linestyle='--', alpha=0.7)

fig.suptitle('Temperatura e Pressão Atmosférica - Transição 2023/2024', 
             fontsize=16, 
             fontweight='bold',
             y=0.98)

plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------------------------------
# Data Concatenation (Visualization in Graph) - Todas as variáveis
# --------------------------------------------------------------------

df_concat_period = df_concat["2023-12-31 20:00:00":"2024-01-01 05:00:00"]
virada_ano = pd.Timestamp("2024-01-01 00:00:00")

# Lista de todas as variáveis para visualização
variaveis = columns_2023

# Cores para 2023 e 2024
cores_2023 = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive',
              'cyan', 'magenta', 'yellow', 'black', 'navy', 'teal', 'coral', 'lime', 'indigo']

cores_2024 = ["#DC9A9A", "#94D394", "#9494D4", "#CCAE90", "#B38AC8", "#BE9D9D", "#C185AD", "#A3A3A3", "#B2C78A",
              "#91CECE", "#BE86BE", "#C6C68B", "#939191", "#9292BD", "#86ABAB", "#BEA0A0", "#A8C58A", "#A385C1"]

# Dividir em grupos de 6 variáveis por figura
variaveis_por_figura = 6
n_figuras = (len(variaveis) + variaveis_por_figura - 1) // variaveis_por_figura

for fig_num in range(n_figuras):
    start_idx = fig_num * variaveis_por_figura
    end_idx = min((fig_num + 1) * variaveis_por_figura, len(variaveis))
    variaveis_grupo = variaveis[start_idx:end_idx]
    
    n_rows_grupo = (len(variaveis_grupo) + 1) // 2
    fig, axes = plt.subplots(n_rows_grupo, 2, figsize=(12, 2 * n_rows_grupo), facecolor='#f9fcffb3')
    axes = axes.flatten()
    
    for i, variavel in enumerate(variaveis_grupo):
        ax = axes[i]
        
        # Dados de 2023
        mask_2023 = df_concat_period.index < virada_ano
        if any(mask_2023):
            ax.plot(df_concat_period.index[mask_2023], 
                    df_concat_period.loc[mask_2023, variavel], 
                    label='2023', 
                    color=cores_2023[start_idx + i], 
                    linewidth=2,
                    marker='o',
                    markersize=3)
        
        # Dados de 2024
        mask_2024 = df_concat_period.index >= virada_ano
        if any(mask_2024):
            ax.plot(df_concat_period.index[mask_2024], 
                    df_concat_period.loc[mask_2024, variavel], 
                    label='2024', 
                    color=cores_2024[start_idx + i], 
                    linewidth=2,
                    marker='s',
                    markersize=3)
        
        ax.set_title(f'{variavel}', fontsize=10, fontweight='bold', pad=5)
        ax.set_ylabel(variavel.split('(')[-1].split(')')[0] if '(' in variavel else 'Valor', fontsize=9)
        ax.legend(loc='best', fontsize=8)
        ax.grid(True, alpha=0.2)
        ax.set_facecolor("#ffffffd0")
        ax.axvline(x=virada_ano, color='gray', linestyle='--', alpha=0.7)
        
        if i >= len(variaveis_grupo) - 2:
            ax.tick_params(axis='x', rotation=45)
        else:
            ax.set_xticklabels([])
    
    # Remover subplots vazios
    for i in range(len(variaveis_grupo), len(axes)):
        fig.delaxes(axes[i])
    
    fig.suptitle(f'Concatenação de Dados - Transição 2023/2024 - Variáveis {start_idx+1}-{end_idx}', 
                 fontsize=14, 
                 fontweight='bold',
                 y=1.05)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
```

```python
# Encontrar o range temporal completo
start_time = df_concat.index.min()
end_time = df_concat.index.max()

frequencia_horaria = 'h'
indice_completo = pd.date_range(start=start_time, end=end_time, freq=frequencia_horaria)

print(f"Índice original: {len(df_concat)} registros")
print(f"Índice completo: {len(indice_completo)} registros")
print(f"Registros faltantes: {len(indice_completo) - len(df_concat)}")
```

```python
# Reindexar o DataFrame para o índice completo
df_concat_reindexed = df_concat.reindex(indice_completo)

print("Dados faltantes após reindexação:")
print(df_concat_reindexed.isnull().sum())
```

```python
# --------------------------------------------------------------------
# Data Concatenation (Visualization in Graph) - Todas as variáveis
# --------------------------------------------------------------------

df_concat_filled_period = df_concat_filled["2023-12-31 20:00:00":"2024-01-01 05:00:00"]
virada_ano = pd.Timestamp("2024-01-01 00:00:00")

# Lista de todas as variáveis para visualização
variaveis = columns_2023

# Cores para 2023 e 2024
cores_2023 = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive',
              'cyan', 'magenta', 'yellow', 'black', 'navy', 'teal', 'coral', 'lime', 'indigo']

cores_2024 = ["#DC9A9A", "#94D394", "#9494D4", "#CCAE90", "#B38AC8", "#BE9D9D", "#C185AD", "#A3A3A3", "#B2C78A",
              "#91CECE", "#BE86BE", "#C6C68B", "#939191", "#9292BD", "#86ABAB", "#BEA0A0", "#A8C58A", "#A385C1"]

# Dividir em grupos de 6 variáveis por figura
variaveis_por_figura = 6
n_figuras = (len(variaveis) + variaveis_por_figura - 1) // variaveis_por_figura

for fig_num in range(n_figuras):
    start_idx = fig_num * variaveis_por_figura
    end_idx = min((fig_num + 1) * variaveis_por_figura, len(variaveis))
    variaveis_grupo = variaveis[start_idx:end_idx]
    
    n_rows_grupo = (len(variaveis_grupo) + 1) // 2
    fig, axes = plt.subplots(n_rows_grupo, 2, figsize=(12, 2 * n_rows_grupo), facecolor='#f9fcffb3')
    axes = axes.flatten()
    
    for i, variavel in enumerate(variaveis_grupo):
        ax = axes[i]
        
        # Dados de 2023
        mask_2023 = df_concat_filled_period.index < virada_ano
        if any(mask_2023):
            ax.plot(df_concat_filled_period.index[mask_2023], 
                    df_concat_filled_period.loc[mask_2023, variavel], 
                    label='2023', 
                    color=cores_2023[start_idx + i], 
                    linewidth=2,
                    marker='o',
                    markersize=3)
        
        # Dados de 2024
        mask_2024 = df_concat_filled_period.index >= virada_ano
        if any(mask_2024):
            ax.plot(df_concat_filled_period.index[mask_2024], 
                    df_concat_filled_period.loc[mask_2024, variavel], 
                    label='2024', 
                    color=cores_2024[start_idx + i], 
                    linewidth=2,
                    marker='s',
                    markersize=3)
        
        ax.set_title(f'{variavel}', fontsize=10, fontweight='bold', pad=5)
        ax.set_ylabel(variavel.split('(')[-1].split(')')[0] if '(' in variavel else 'Valor', fontsize=9)
        ax.legend(loc='best', fontsize=8)
        ax.grid(True, alpha=0.2)
        ax.set_facecolor("#ffffffd0")
        ax.axvline(x=virada_ano, color='gray', linestyle='--', alpha=0.7)
        
        if i >= len(variaveis_grupo) - 2:
            ax.tick_params(axis='x', rotation=45)
        else:
            ax.set_xticklabels([])
    
    # Remover subplots vazios
    for i in range(len(variaveis_grupo), len(axes)):
        fig.delaxes(axes[i])
    
    fig.suptitle(f'Concatenação de Dados - Transição 2023/2024 - Variáveis {start_idx+1}-{end_idx}', 
                 fontsize=14, 
                 fontweight='bold',
                 y=1.05)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
```
